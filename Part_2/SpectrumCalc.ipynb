```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
nH = 0.1  
x = 0.1
G = 6.265e8  
f = 0.4164
g0 = 2
Z = 2.0
me = 9.11e-28 
c = 3e10  
e = 4.80e-10  
pi = 3.14159265
```

```python
def emissionAlpha(v, v0):
  return (e*e)*f*nH*(1-x)*g0*G/(4*pi*me*c*Z*( ( (v-v0)**2 ) + ( (G/(4*pi))**2 ) ))
def emissionIntensity(v,v0,d):
  return np.exp(-emissionAlpha(v,v0)*d)
```

```python
def plotFreqsRange(wl,wh,d):
  wavelengths = []
  intensities = []
  for wavelength in range(wl,wh):
    wavelengths.append(wavelength)
    intensities.append(emissionIntensity(c/(wavelength/10**7),v0,d))
  plt.plot(wavelengths,intensities)
  plt.xlabel(("Wavelength"))
  plt.ylabel(("Intensity"))
  plt.title("Intensity Spectrum")
  print("Wavelength for minimum intensity is: ",wavelengths[np.argmin(intensities)])
```

### First Set of values

```python
v0 = 2.46607e15
d = 10**14
plotFreqsRange(80,160,d)
```

```python
v0 = 2.46607e15
d = 10**18
plotFreqsRange(80,160,d)
```

```python
v0 = 2.46607e15
d = 10**21
plotFreqsRange(80,160,d)
```

### Second Set of values

```python
v0 = 2.46632e15
d = 10**14
plotFreqsRange(80,160,d)
```

```python
v0 = 2.46632e15
d = 10**21
plotFreqsRange(80,160,d)
```

# Conclusions
 
### Similarities

* For all spectra we have minimum intensity wavelength at 122nm.
* On improving resolution perhaps we could get a better estimate.
* For the Lyman Transition, to n = 2, the actual value is 121.57nm.
* Notice the existence of square relation in v-v0 makes the graph symmetric around the minima intensity wavelength.

### Differences

* On simulations with finer gaps between wavelengths, the second set has lower wavelength for intensity minima due to shift in v0.
* For spectra with thicker slabs, more wavelengths appear to be absorbed and the minima is less sharp.
